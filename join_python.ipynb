```python
import numpy as np
import pandas as pd
```

```python
#import the department table
dept = pd.read_csv('department.csv')
#import the employee table
emp = pd.read_csv('employees.csv')
```

```python
dept.head()
```

```python
# round to two decimal places in python pandas
pd.options.display.float_format = '{:.0f}'.format
```

```python
emp.head()
```

```python
#Setting sort=True arranges the results by department_id
emp.merge(dept, on = "department_id",sort=True)
```

```python
#Setting sort=True arranges the results by department_id
emp.merge(dept, how = 'left', on = "department_id",sort=True)
```

```python
emp.merge(dept, how = 'right', on = "department_id",sort=True)
```

```python
emp.merge(dept, how = 'outer', on = "department_id",sort=True)
```

|               | SQL JOINs                                     | Python JOINs                                          |
|-------------------------|-----------------------------------------------|-------------------------------------------------------|
| Data Source             | Tables in relational databases               | Operates on in-memory data structures (dataframes)   |
| Joining Method          | SQL syntax using JOIN keyword                | `.merge()` method                                     |
| Criteria Matching       | ON or USING clause                           | on or left_on or right_on parameters                  |
| Types of Joins          | INNER, LEFT, RIGHT, OUTER                       | INNER, LEFT, RIGHT, OUTER                        |
| Default Join Type       | INNER                                         | INNER                                                 |
| Additional Parameters   | WHERE, ORDER BY clauses etc.                 | left_index, right_index, suffixes, sort, validate, etc.|
| Result Set              | Output of the JOIN operation                 | New dataframe containing merged data                 |
| Additional Clauses      | Queries may become cumbersome with extra clauses | Parameters and arguments make it easy               |
| Performance             | Efficient for large datasets                 | Less efficient for data that can't fit into memory   |
| Portability             | Queries are portable across database systems | Only in Python environment                           |
